```python
# if selenium is used for the first time -- use -- pip install selenium
#its a package used for automation

from selenium import webdriver
# To interact with chrome webdriver is used

from selenium.webdriver.common.by import By
# By module is imported to acsess search button

browser = webdriver.Chrome(executable_path="C:/chromedriver/chromedriver")
# To open chrome window, an instance is made
```

```python
category_names = browser.find_elements(By.XPATH,"//ul[contains(@class, 'nav category-bar')]")
categories = ""
for product in category_names:
    categories = categories + product.text
print(categories)
categories = categories.split('\n')
print(categories)
```

```python
def item_details(i):
    product_names = browser.find_elements(By.XPATH,"//span[contains(@class, 'product-title')]")
    print("PRODUCTS IN THIS SUB CATEGORY ARE:")   
    print('\n')
    for product in product_names:
        print(product.text)
        i=i+1
    print('\n')
    print("NUMBER OF PRODUCTS IN THIS SUB CATEGORY IS:",i)
```

```python
def item_number(j):
    product_names = browser.find_elements(By.XPATH,"//span[contains(@class, 'product-title')]")    
    for product in product_names:
        j=j+1
    return j
```
